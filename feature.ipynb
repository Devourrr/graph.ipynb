```python
import requests
import pandas as pd

df_total = pd.DataFrame()
url = 'http://apis.data.go.kr/B552061/lgStat/getRestLgStat'

for year in range(2010,2021):
    print(f"Year: {year}")
    
    params = {
    "ServiceKey" : 'HCnD0rtW2xL5JgkUNzOEFPuVQO9FxnqdTxFgkAkNYh2eacEW62WkpsWVn7W+UoQNKJXa1HIc8IR9t6ZkI4wzIA=='
    , 'searchYearCd': year
    , 'siDo': 1100 # 서울 전체
    , 'guGun': '' # 서울 전체 구
    , 'type': 'json'
    , 'numOfRows': 5000
    , 'pageNo': 1
    }
    
    res = requests.get(url, params = params)
#     print(res.json())
    
    data = res.json()['items']['item']
    df = pd.DataFrame(data=data)
    
    cond = df['acc_cl_nm'] != '전체사고'
    df = df[cond]
    df = df.iloc[:, :13].copy()
    
    df['std_year'] = df['std_year'].astype('int')
    df['acc_cnt'] = df['acc_cnt'].astype('int')
    df['acc_cnt_cmrt'] = df['acc_cnt_cmrt'].astype('float')
    df['dth_dnv_cnt'] = df['dth_dnv_cnt'].astype('int')
    df['dth_dnv_cnt_cmrt'] = df['dth_dnv_cnt_cmrt'].astype('float')
    df['ftlt_rate'] = df['ftlt_rate'].astype('float')
    df['injpsn_cnt'] = df['injpsn_cnt'].astype('int')
    df['injpsn_cnt_cmrt'] = df['injpsn_cnt_cmrt'].astype('float')
    df['tot_acc_cnt'] = df['tot_acc_cnt'].astype('int')
    df['tot_dth_dnv_cnt'] = df['tot_dth_dnv_cnt'].astype('int')
    df['tot_injpsn_cnt'] = df['tot_injpsn_cnt'].astype('int')
    
    df_total = df_total.append(df)
#     print(df_total)
df_total = df_total.set_index('std_year')
df_total
    
```

```python
def sigu(x):
    y = len(x.split())
    return y
#함수를 적용하는 함수 apply(함수명), 사용자 정의 함수 사용 가능
cond = df_total['sido_sgg_nm'].apply(sigu) == 1
print(cond)
```
